```
import requests
from bs4 import BeautifulSoup
import os
import csv
import pandas as pd

# Lấy nội dung từ các bài báo đã crawl từ bài tập trước 
file = open("/content/drive/My Drive/My_Data/Data_crawl_from_Tuoi_Tre/data.csv")
reader = csv.reader(file)
header = next(reader)
# Check file as empty
if header != None:
  for i,line in enumerate(reader):
    print("Processing file {}........... ".format(i))
    news = requests.get(line[2]) #line[2] is a url of news
    soup = BeautifulSoup(news.content, "html.parser")
    try:
      body = soup.find("div", id="main-detail-body")
    except:
      body = "Format web not similar pre-page "
    try:
      content = body.findChildren("p", recursive=False)[0].text +      body.findChildren("p", recursive=False)[1].text
    except:
      content = "Not Exit"
    
    file_name = "News_" + str(i) + ".txt"
    path_all_news = "/content/drive/My Drive/My_Data/Data_crawl_from_Tuoi_Tre/Danh sách các bài báo"
    path_save = os.path.join(path_all_news,file_name)
    
    text_file = open(path_save, "w")
    text_file.write(content)
    text_file.close()
    print("Done file {}".format(i))
# Lưu các nội dung của từng bài viết thành danh sách các file 
```

Đoạn code dưới đây chạy từ đầu tới cuối, còn đoạn ở trên phân ra từng bước 

```
import glob
import numpy as np
import re
import requests
from bs4 import BeautifulSoup
import os
import csv



def Build_Dictionary(file_paths):
  lst_contents = []
  dictionary = set()
  dem_test = 0
  dic_file = {}
  for i,file_path in enumerate(file_paths):
    dic_file[i] = file_path.split("/")[-1]
    if dem_test < 11:
      print("Processing {}..........".format(file_path.split("/")[-1]))
      f = open(file_path, 'r')
      str = f.read().lower()
      # lst_contents.append(str)
      
      words = str.replace('"', '').replace('.', '').replace("'","").replace("(", "").split()
      
      words = set(words)
      lst_contents.append(words)
      dictionary.update(words)
    dem_test += 1
    #print(words)
    #print ('Noi dung file cua ban la:\n', str)
  print('Tap cac tu: ', dictionary)
  dictionary = list(dictionary)
  return dictionary, lst_contents, dic_file

def Build_Inverted_Files(dictionary,lst_contents):
  inv_files = dict()
  all_file = set()
  for k,word in enumerate(dictionary):
      inv_files[word] = set()
      for i,content in enumerate(lst_contents):
          if word in content:
              inv_files[word].add(i)
              #print(word,"    ",content)
          all_file.add(i)
  inv_files['all_file'] = all_file
  inv_files['unknow'] = set()
  print('inv_files',inv_files)
  return inv_files


def Split_Query(query):
  query = query.strip(" ")
  querys = re.sub('\s+',' ',query).split(" ")
  print(querys)
  query_words = []
  query_logical = []
  logicals = ["and","not", "or", "xor", "nor"]
  for i in range(len(querys)):
    if querys[i] not in logicals:
      query_words.append(querys[i])
    else:
      query_logical.append(querys[i])
  print("query_words: ",query_words)
  print("query_logical: ",query_logical)
  return query_words, query_logical

  # BUOC 4: Tra từng từ trong inverted files
  # để lấy tập hợp văn bản có chứa từ
def Processing_Boolean_Querie(query_words,inv_files):
  results = []
  for query_word in query_words:
      # tra vi tri cua query_word trong dictionary
      S = set()
      try:
          S = inv_files[query_word]
      except:
          S = inv_files['unknow']
      print('Tap hop van ban chua tu ', query_word, ' la: ', S)
      results.append(S)
  return results


# BUOC 5: thực hiện phép biến đổi luận lý: AND, OR, XOR, NOT...


def Compute_Logical(query_words,query_logical,inverted_words, inv_files):

  result = inverted_words[0]
  ind_log = 0 
  for i in range(len(inverted_words)):
    i_temp = ind_log + 1
    if i_temp  >= len(query_logical) or query_logical[i_temp] != "not":
      b = inverted_words[i]
    else:
      b = inverted_words[i]
      while( i_temp < len(query_logical) and query_logical[i_temp] == "not"):
        b = inv_files['all_file'] - b
        #print('b',b)
        i_temp += 1
    
    #print("ind_log",ind_log)
    if query_logical[ind_log] == "and":
      result = result & b
    elif query_logical[ind_log] == "or":
      result = result | b
    elif query_logical[ind_log] == "xor":
      result = result ^ b
    elif query_logical[ind_log] == "nor":
      result = inv_files['all_file'] - (result | b)
    ind_log = i_temp - 1
  return result
      


def show_result(result,dic_file):
  if result != set():
    data = pd.read_csv("/content/drive/My Drive/My_Data/Data_crawl_from_Tuoi_Tre/data.csv")
    for file_name in result:
      num_line = int(dic_file[file_name].split("_")[-1].split(".")[0])
      print(data.iloc[num_line])
  else:
    print("Not found")



def main():
  # BUOC 1: Đọc file text trong thư mục 'data'
  print("______________________Step 1: Đọc file text trong thư mục 'Data_crawl_from_Tuoi_Tre'____________\n")
  # va xay dung tap tu dien
  # Lay danh sach file txt trong thu muc data
  file_paths = glob.glob("/content/drive/My Drive/My_Data/Data_crawl_from_Tuoi_Tre/Danh sách các bài báo/*.txt")
  #print(txt_files)

  # BUOC 2: Xây dựng dictionary
  print("_______________________Step 2: Build dictionary______________________________________________\n")
  # Doc noi dung cua tung file txt
  # va xay dung tap "dictionary" chua danh sach cac tu
  dictionary,lst_contents, dic_file = Build_Dictionary(file_paths)

  # BUOC3: Xây dụng invited file
  print("_______________________Step 3: Build inveted file_____________________________________________v")
  inv_files = Build_Inverted_Files(dictionary,lst_contents)
  print("all file",inv_files["all_file"])
  # BUOC 4: Nhập câu truy vấn và phân tách từ truy vấn
  print("_______________________Step 4: Nhập câu truy vấn và phân tích câu truy vấn____________________\n")
  #query = input("Mơi bạn nhập câu cần truy vấn: ")
  query = " dựng or not not   Chất and not Trí or Nhật "
  query = query.lower()
  # Phân tác câu truy vấn
  query_words,query_logical = Split_Query(query)

  # BUOC5: Tra từ trong inverted file
  print("________________________Step 5: search in inverted file__________________________________________\n")
  inverted_words = Processing_Boolean_Querie(query_words,inv_files)

  print("inverted_words: ", inverted_words)
  print("________________________Step 6: Compute logical_________________________________________________\n")
  # BUOC 6: Thực hiện các phép tính logic AND, OR, NOT,...
  result = Compute_Logical(query_words,query_logical, inverted_words, inv_files)
  print("result",result)
  # BUOC 7: Tra về kết quả
  print("_________________________Step 7: Result_____________________________________________________________\n")
  show_result(result,dic_file)

if __name__ == "__main__":
  main()

```
